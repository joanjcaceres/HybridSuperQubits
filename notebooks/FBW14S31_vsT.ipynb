```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
T = np.array([22, 36, 100, 200, 300, 400, 475])

f0_M21_off = np.array([4.8532, 4.8532, 4.8532, 4.8531, 4.852, 4.850, 4.843])
f0_M22_off = np.array([4.704, 4.704, 4.704, 4.704, 4.702, 4.699, 4.687])

Ki_M21_off = np.array([0.142, 0.142, 0.161, 0.289, 1.107, 2.38, 3.63])
Ki_M22_off = np.array([0.828, 0.863, 0.916, 1.08, 2.43, 4.34, 7.95])

Kc_M21_off = np.array([0.331, 0.332, 0.346, 0.358, 0.435, 0.477, 0.362])
Kc_M22_off = np.array([6.64, 6.7, 6.62, 6.65, 6.81, 6.85, 6.80])

Qi_M21_off = f0_M21_off*1e3 / Ki_M21_off
Qi_M22_off = f0_M22_off*1e3 / Ki_M22_off

Qc_M21_off = f0_M21_off*1e3 / Kc_M21_off
Qc_M22_off = f0_M22_off*1e3 / Kc_M22_off

# Crear los grficos
fig, axs = plt.subplots(5, 1, figsize=(10, 25))

# Grfico de f0 vs T
axs[0].plot(T, f0_M21_off, 'b--', label='M21 TWPA Off')
axs[0].plot(T, f0_M22_off, 'r--', label='M22 TWPA Off')
axs[0].set_xlabel('T (mK)')
axs[0].set_ylabel('f0 (GHz)')
axs[0].set_title('f0 vs T')
axs[0].legend()

# Grfico de Ki vs T
axs[1].plot(T, Ki_M21_off, 'b--', label='M21 TWPA Off')
axs[1].plot(T, Ki_M22_off, 'r--', label='M22 TWPA Off')
axs[1].set_xlabel('T (mK)')
axs[1].set_ylabel('$Ki/2\pi$ (MHz)')
axs[1].set_title('Ki vs T')
axs[1].legend()

# Grfico de Kc vs T
axs[2].plot(T, Kc_M21_off, 'b--', label='M21 TWPA Off')
axs[2].plot(T, Kc_M22_off, 'r--', label='M22 TWPA Off')
axs[2].set_xlabel('T (mK)')
axs[2].set_ylabel('$Kc/2\pi$ (MHz)')
axs[2].set_title('Kc vs T')
axs[2].legend()

# Grfico de Q_intrinsic vs T
axs[3].plot(T, Qi_M21_off, 'b--', label='M21 TWPA Off')
axs[3].plot(T, Qi_M22_off, 'r--', label='M22 TWPA Off')
axs[3].set_xlabel('T (mK)')
axs[3].set_ylabel('Q_intrinsic')
axs[3].set_title('Q_intrinsic vs T')
axs[3].legend()

# Grfico de Q_coupling vs T
axs[4].plot(T, Qc_M21_off, 'b--', label='M21 TWPA Off')
axs[4].plot(T, Qc_M22_off, 'r--', label='M22 TWPA Off')
axs[4].set_xlabel('T (mK)')
axs[4].set_ylabel('Q_coupling')
axs[4].set_title('Q_coupling vs T')
axs[4].legend()


plt.tight_layout()
plt.show()
```
