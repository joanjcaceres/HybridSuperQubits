```python
def eigensystem_andreev(EDelta,phi_ext,r, eigvals = 0):
    # EIGENVALUES AND EIGENSTATES OF THE FERMIONIC-BOSONIC QUBIT 

    ReZ = (np.cos(phi_ext/2)*np.cos(r*phi_ext/2)+r*np.sin(phi_ext/2)*np.sin(r*phi_ext/2)) #Re(Z) of the Hamiltonian
    ImZ = (-np.cos(phi_ext/2)*np.sin(r*phi_ext/2)+r*np.sin(phi_ext/2)*np.cos(r*phi_ext/2)) ##Im(Z) of the Hamiltonian
    H = EDelta*(ReZ*sigmaz()+ImZ*sigmay()) #Hamiltonian.
    evals,ekets=H.eigenstates(eigvals=eigvals)

    dReZdr = 1/2*(r*phi_ext*np.cos(r*phi_ext/2)*np.sin(phi_ext/2)+(-phi_ext*np.cos(phi_ext/2)+2*np.sin(phi_ext/2))*np.sin(r*phi_ext/2))
    dImZdr = -1/2*np.cos(r*phi_ext/2)*(phi_ext*np.cos(phi_ext/2)-2*np.sin(phi_ext/2))-1/2*r*phi_ext*np.sin(phi_ext/2)*np.sin(r*phi_ext/2)
    dHdr = EDelta*(dReZdr*sigmaz()+dImZdr*sigmay())
    dHdr_me = dHdr.matrix_element(ekets[1],ekets[0]) 
    return evals,ekets,dHdr_me
```
